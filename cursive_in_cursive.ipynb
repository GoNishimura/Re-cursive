```python
def move_points(array, x_pad=0, y_pad=0, scale=1):
    if type(array) != np.ndarray: array = np.array(array).astype(np.double) 
#     in_array = array#np.array(array).astype(np.double)
#     x_array = (array[0] + x_pad) * scale
#     y_array = (array[1] + y_pad) * scale
    x_array = array[0] * scale + x_pad 
    y_array = array[1] * scale + y_pad
    
    return np.asfortranarray([x_array, y_array]).astype(np.double)
```

```python
def plot_points(array):
    if type(array) != np.ndarray: array = np.array(array).astype(np.double) 
#     in_array = array#np.array(array).astype(np.double) 
    curves = [bezier.Curve(array[:, i:i+4], degree=2) for i in range(0, array.size, 3)] # in_array.shape[1]-1

    fig, ax = plt.subplots()
    for cu in curves:
        cu.plot(num_pts=256, ax=ax, color='k')
#     ax.set_xlim(0, 60)
#     ax.set_ylim(0, 66)
#     ax.set_xlim(-60, 60)
#     ax.set_ylim(-60, 60)
    ax.set_aspect("equal")
    plt.show()
```

```python
def stack_points(list_points, plot=(4, 4)):
    if type(list_points) == str:
        list_points = [to_points(char) for char in list_points]
        
    curves = []
    end_points = np.array([[],[]])
    for i, points in enumerate(list_points):
        #print(points)
        strt_points = points[:, :4]
        mid_points = points[:, 3:-3]
        #print(mid_points)
        
        to_begin = np.hstack([end_points, strt_points])
        curves += [bezier.Curve(to_begin, degree=2)]
        curves += [bezier.Curve(mid_points[:, i:i+4], degree=2) for i in range(0, mid_points[0].size, 3) if mid_points[0, i:i+4].size==4]
        end_points = points[:, -4:]
        if i == len(list_points)-1:
            curves += [bezier.Curve(end_points, degree=2)]
    
    if plot == False: return curves
    
    fig, ax = plt.subplots()
    for cu in curves:
        cu.plot(num_pts=256, ax=ax, color='k')
    ax.axis('off')
    ax.set_xlim(-25, 140)
    ax.set_ylim(-25, 20)
    ax.set_aspect("equal")
    plt.savefig('sp.pdf')
#     plt.show()
```

```python
def cic(meta_chars, base_chars='CCCCCC', scale=0.1, plot=(4, 4)):
    meta_curves = stack_points(meta_chars, False) if type(meta_chars) == str else meta_chars
    splitter = np.linspace(0, 1, len(base_chars))
    
    new_curves = []
    for cu in meta_curves:
        centers = cu.evaluate_multi(splitter)
        for i, (center_x, center_y) in enumerate(zip(centers[0], centers[1])): 
            new_points = move_points(to_points(base_chars[i]), center_x, center_y, scale)
            new_curves.append(new_points)
            
    if plot == False: return stack_points(new_curves, False)
        
    fig, ax = plt.subplots(figsize=plot)
    for cu in stack_points(new_curves, False):
        cu.plot(num_pts=256, ax=ax, color='k')
    ax.set_xlim(-25, 140)
    ax.set_ylim(-25, 20)
    ax.axis('off')
    ax.set_aspect("equal")
    plt.savefig('cic{}.pdf'.format(plot[0]*plot[1]))
#     plt.show()
```

```python
def phrase2points(phrase, scale=1):
    points = []
    half_word_size = 20*scale
    
    dist = 0
    before_lower = True
    for i, c in enumerate(phrase):
        if i != 0:
            dist += half_word_size if not before_lower else half_word_size/2
        points.append(move_points(to_points(c), dist, 0, scale))
        before_lower = c.islower()
        dist += half_word_size if not before_lower else half_word_size/2
        
    return points
```
